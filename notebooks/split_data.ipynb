```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Amaan\\Desktop\\supply-chain-analytics-sql\\data\\DataCoSupplyChainDataset.csv", encoding="latin1")
df.head(5)
```

```python
# create dimension customers
customers = df[[
    'Customer Id',
    'Customer Fname',
    'Customer Lname',
    'Customer Email',
    'Customer Segment',
    'Customer City',
    'Customer State',
    'Customer Country',
    'Customer Zipcode',
    'Customer Street'
]].drop_duplicates()

customers.to_csv(
    'C:\\Users\\Amaan\\Desktop\\supply-chain-analytics-sql\\data\\processed\\dim_customers.csv',
    index=False
)
```

```python
# create dimension product
products = df[[
    'Product Card Id',
    'Product Name',
    'Product Price',
    'Product Status',
    'Category Id',
]].drop_duplicates()

products.to_csv(
    'C:\\Users\\Amaan\\Desktop\\supply-chain-analytics-sql\\data\\processed\\dim_products.csv',
    index=False
)
```

```python
# create dimension category
category = df[[
    'Category Id',
    'Category Name',
    'Department Id',
]].drop_duplicates()

category.to_csv(
    'C:\\Users\\Amaan\\Desktop\\supply-chain-analytics-sql\\data\\processed\\dim_categories.csv',
    index=False
)
```

```python
# create dimension department
department = df[[
    'Department Id',
    'Department Name',
]].drop_duplicates()

department.to_csv(
    'C:\\Users\\Amaan\\Desktop\\supply-chain-analytics-sql\\data\\processed\\dim_departments.csv',
    index=False
)
```

```python
# create dimension georaphy
geography = df[[
    'Market',
    'Order Region',
    'Order Country',
    'Order State',
    'Order City',
    'Order Zipcode',
    'Latitude',
    'Longitude'
]].drop_duplicates()

geography.reset_index(drop=True, inplace=True)
geography.insert(0, 'geo_id', geography.index + 1)

geography.to_csv(
    'C:\\Users\\Amaan\\Desktop\\supply-chain-analytics-sql\\data\\processed\\dim_geography.csv',
    index=False
)
```

```python
# Now add the surrogate PK of geography in the main table for reference in Orders fact table
geography_columns = [
"Market",
"Order Region",
"Order Country",
"Order State",
"Order City",
"Order Zipcode",
"Latitude",
"Longitude"
]

df = df.merge(geography, on=geography_columns, how="left")
```

```python
# create main fact table orders
orders = df[[
    'Order Item Id',
    'Order Id',
    'Customer Id',
    'Product Card Id',
    'geo_id',
    'order date (DateOrders)',
    'Order Item Quantity',
    'Sales',
    'Order Item Total',
    'Benefit per order',
    'Order Profit Per Order',
    'Order Item Profit Ratio',
    'Order Status'
]]

orders.to_csv(
    'C:\\Users\\Amaan\\Desktop\\supply-chain-analytics-sql\\data\\processed\\fact_orders.csv',
    index=False
)
```

```python
# create shipping fact
shipping = df[[
    'Order Id',
    'shipping date (DateOrders)',
    'Days for shipping (real)',
    'Days for shipment (scheduled)',
    'Delivery Status',
    'Late_delivery_risk',
    'Shipping Mode'
]]

shipping.to_csv(
    'C:\\Users\\Amaan\\Desktop\\supply-chain-analytics-sql\\data\\processed\\fact_shipping.csv',
    index=False
)
```
